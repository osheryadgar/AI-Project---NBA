```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from prettytable import PrettyTable
from matplotlib import pyplot
from sklearn.cluster import KMeans
pd.set_option('display.max_columns', None)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('all_seasons.csv')
df.head(10)
```

```python
df_dummies = pd.get_dummies(df, columns=['team_abbreviation', 'college', 'country'])
df_dummies = df_dummies.drop('draft_round', axis=1)
df_dummies.head()
```

```python
df_dummies['draft_year'] = df_dummies['draft_year'].replace('Undrafted', np.nan)
df_dummies['draft_number'] = df_dummies['draft_number'].replace('Undrafted', np.nan)
df_dummies[['draft_year', 'draft_number']].head(10)
```

```python
df_dummies.isnull().sum().head(17)
```

```python
df_dummies = df_dummies.dropna()
```

```python
X = df_dummies.iloc[:,list(range(2,4)) + list(range(6,7)) + list(range(8,16))]
X.head()
```

```python
y = df_dummies['pts']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# test - train
# 20% - 80%
```

```python
X_train.head()
```

```python
# Scale the training set features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Scale the test set features using the scaler fit to the training set features
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a LinearRegression model and fit it to the training data
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
predictions = model.predict(X_test_scaled)
```

```python
# Evaluate the model's performance on the test set
score = model.score(X_test_scaled, y_test)
print(score)
```

# K-Means

```python
full = df_dummies
full = full.dropna()
```

```python
removed_cols = (list(full.columns[full.dtypes == "object"]) + ['target'])
selected_cols = full.columns[~full.columns.isin(removed_cols)]

kmeans = KMeans(5)
kmeans.fit(full[selected_cols])
pd.DataFrame(kmeans.cluster_centers_, columns=selected_cols).T
```

```python
full
```

```python
full['grade'] = kmeans.labels_
full['grade'].value_counts()
```

```python
def add_target2(group):
    group["target"] = group["grade"].shift(-1)
    group["target"].fillna(5, inplace=True)
    return group
```

```python
full = full.groupby("player_name", group_keys=False).apply(add_target2)
```

```python
full[full['player_name'] == 'Deni Avdija']['grade'].value_counts()
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit


def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)



rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)
sfs = SequentialFeatureSelector(rr,
                                n_features_to_select=30, 
                                direction="forward",
                                cv=split,
                                n_jobs=1
                               )
```

```python
selected_columns = X.columns
selected_columns
```

```python
#sfs.fit(full[selected_columns], full["target"])
#predictors = list(selected_columns[sfs.get_support()])

predictions = backtest(full, rr, selected_columns)
accuracy_score(predictions["actual"], predictions["prediction"])
```

```python
predictions['diff'] = predictions['prediction'] - predictions['actual']
```

```python
predictions['diff'].describe()
```

```python
ax = predictions['diff'].plot(kind='hist')

# calculate the mean and standard deviation of the 'diff' column
mean = predictions['diff'].mean()
std = predictions['diff'].std()

# add vertical lines for the mean and +/- 1 standard deviation
ax.axvline(mean, color='r', linestyle='dashed', linewidth=1)
ax.axvline(mean - std, color='g', linestyle='dashed', linewidth=1)
ax.axvline(mean + std, color='g', linestyle='dashed', linewidth=1)

# set the plot title and axis labels
ax.set_title('Histogram of Differences')
ax.set_xlabel('Difference')
ax.set_ylabel('Frequency')

# show the plot
plt.show()
```
