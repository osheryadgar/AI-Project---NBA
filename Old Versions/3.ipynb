```python
df_dummies = pd.get_dummies(df, columns=['team_abbreviation', 'college', 'country'])
df_dummies = df_dummies.drop('draft_round', axis=1)
df_dummies.head()
```

```python
df_dummies['draft_year'] = df_dummies['draft_year'].replace('Undrafted', np.nan)
df_dummies['draft_number'] = df_dummies['draft_number'].replace('Undrafted', np.nan)
df_dummies[['draft_year', 'draft_number']].head(10)
```

```python
df_dummies.isnull().sum().head(17)
```

```python
df_dummies[df_dummies['player_name'] == 'Deni Avdija']
```

```python
df_dummies = df_dummies.dropna()
columns = ['age', 'player_height', 'pts', 'reb', 'ast', 'net_rating', 'oreb_pct', 'dreb_pct', 'usg_pct', 'ts_pct']
X = df_dummies[columns]

X_scaled = (X - X.mean()) / X.std()
X_scaled
```

```python
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

# Create a new column with the composite scores
df_dummies['score'] = X_pca
```

```python
data = df_dummies.sort_values(by='score', ascending=False)
data[['player_name'] + columns + ['score']]
```

```python
df_dummies['game_score'] = df_dummies['pts'] + 0.4*df_dummies['FG'] - 0.7*df_dummies['FGA'] - 0.4*(df_dummies['FTA'] - df_dummies['FT']) + 0.7*df_dummies['ORB'] + 0.3*df_dummies['DRB'] + df_dummies['STL'] + 0.7*df_dummies['AST'] + 0.7*df_dummies['BLK'] - 0.4*df_dummies['PF'] - df_dummies['TOV']
```
